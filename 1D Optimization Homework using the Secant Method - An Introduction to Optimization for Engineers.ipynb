```julia
function zeroth_derivative(x)
    
    return sin(x) + x - ((x^2)/10)
    
end
```

```julia
function first_derivative(x)
    
    return cos(x) + 1 - (x/5)
    
end
```

```julia
function second_derivative(x)
    
    return -sin(x)-(1/5)
    
end
```

```julia
x = 4
dx = 0.01

for iteration = 1:100
    
    numerator = (zeroth_derivative(x+dx) - zeroth_derivative(x-dx))/(2*dx)
    
    denominator = (zeroth_derivative(x+dx) - 2*zeroth_derivative(x) + zeroth_derivative(x-dx))/(dx^2)
    
    x = x - numerator/denominator
    
end
```
